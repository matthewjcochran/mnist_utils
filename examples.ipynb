```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

from mnist_utils import show_sample_grid, get_predictions, get_titles
```

# MNIST Dataset

```python
root="./data"
transform = transforms.ToTensor()

train_dataset = datasets.MNIST(
    root=root,
    train=True,
    download=True,
    transform=transform
)

test_dataset = datasets.MNIST(
    root=root,
    train=False,
    download=True,
    transform=transform
)
```

```python
print(len(train_dataset))
print(len(test_dataset))
```

```python
loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
images, labels = next(iter(loader))
```

# Show Sample Grid

```python
show_sample_grid(images, labels.tolist())
```

# Show Sample Grid With Predictions

```python
model = nn.Sequential(
    nn.Flatten(),
    nn.Linear(28*28, 10)
)
```

```python
predictions = get_predictions(model, images)
titles = get_titles(labels, predictions)

show_sample_grid(images, titles)
```
